# 4. Which neighorhood group has the highest average price?
## Answer: Manhattan at $196.88

```python
price_by_neighbourhood = airbnb[['neighbourhood_group','price']]
price_by_neighbourhood.groupby('neighbourhood_group').mean().sort_values(by='price',ascending=False).round(2)
```

# 5. Which neighbor hood group has the highest total price?
## Answer: Brooklyn, Manhattan, and Queens all have at least 1 listing for $10,000

```python
price_by_neighbourhood = airbnb[['neighbourhood_group','price']]
price_by_neighbourhood.groupby('neighbourhood_group').max().sort_values(by='price', ascending=False)
```

```python
# or Manhattan, if you're looking for the total price of all their listings, 
# but that's an odd way to look at it

price_by_neighbourhood.groupby('neighbourhood_group').sum().sort_values(by='price', ascending=False)
```

# 7. Who currently has no (zero) availability with a review count of 100 or more?

## 136 listsings or 162 depending on how you look at it

```python
avail = airbnb[['host_id', 'host_name','number_of_reviews','availability_365']]
allSum = avail.groupby('host_id')[['number_of_reviews', 'availability_365']].sum()
answer = allSum[(allSum['availability_365'] == 0) & (allSum['number_of_reviews'] >= 100)]
answer
```

```python
ahundred_reviews = airbnb[airbnb['number_of_reviews'] >= 100]
availability = ahundred_reviews[['id', 'host_id', 'host_name', 'number_of_reviews', 'availability_365']]
availability[availability['availability_365'] == 0].sort_values(by='number_of_reviews', ascending=False)
```

# 8. What host has the highest total of prices and where are they located?
